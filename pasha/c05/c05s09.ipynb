```python
import matplotlib.pyplot as plt

layout = """
AAABB
CCCCC
"""

fig, ax = plt.subplot_mosaic(layout, figsize=(12, 7))
```

```python
new_layout = """
AAABB
C..DD
"""

fig, ax = plt.subplot_mosaic(new_layout, figsize=(12, 7))
```

```python
astro_layout = """
AABBCC
AABBCC
DDDDEE
DDDDEE
DDDDGG
DDDDGG
"""

fig, ax = plt.subplot_mosaic(astro_layout, figsize=(12,12))

for i in ax.values():
    i.set_xticks([])
    i.set_yticks([])
```
