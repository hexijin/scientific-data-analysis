```python
import numpy as np
import matplotlib.pyplot as plt


# change the font globally, then re-plot

plt.rc("font", size=16, family='serif')

# old stuff
x = np.linspace(0, 10, 100)
y = np.random.random(size=len(x))

fig, ax = plt.subplots(figsize=(6,6))

ax.plot(x, y, '.', label='Some Data')
ax.legend(loc=2)
ax.set_xlabel('Time (s)')
ax.set_ylabel('Noise Fluctuations (arb)')
ax.set_xlim(-1, 11)
ax.set_ylim(-0.1, 1.2)

ax.tick_params(axis='both',
              direction='in',
              length=6,
              top=True,
              right=True)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6,6))

ax.plot(x, y, '.', label='Some Data')
ax.set_xlim(-1, 11)
ax.set_ylim(-0.1, 1.2)

# do some LaTeX

ax.set_xlabel(r'Time ($s_{since\ the\ Big\ Bang}$)', fontsize=12)
ax.set_ylabel(r'Noise Fluctuations ($x^{15}$)', fontsize=12)
ax.tick_params(labelsize=12)
ax.legend(loc=1,prop={'size':12})
plt.show()
```
