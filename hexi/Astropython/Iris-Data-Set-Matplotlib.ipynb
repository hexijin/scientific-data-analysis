```python
#making a histogram 

import matplotlib.pyplot as plt
import numpy as np
import csv
from matplotlib.ticker import MultipleLocator

plt.rc("font",size=16,family="serif")

with open('C:/Users/Jin/Downloads/iris_dataset.csv') as f:
    data = [row for row in csv.reader(f) if row]
    sepal_length = [float(row[0]) for row in data]
    sepal_width = [float(row[1]) for row in data]
    petal_length = [float(row[2]) for row in data]
    petal_width = [float(row[3]) for row in data]
    class_of_the_iris_plant = [row[4] for row in data]

sepal_length = np.array(sepal_length)
class_of_the_iris_plant = np.array(class_of_the_iris_plant)

sepal_Setosa = sepal_length[class_of_the_iris_plant=='Iris-setosa']
sepal_Versicolor = sepal_length[class_of_the_iris_plant=='Iris-versicolor']
sepal_Virginica = sepal_length[class_of_the_iris_plant=='Iris-virginica']

fig, ax = plt.subplots(figsize=(10,6))
ax.hist([sepal_Setosa,sepal_Versicolor,sepal_Virginica],bins=4, label=['Setosa','Versicolor','Virginica'])
ax.legend(loc=1)
ax.set_xlabel('Sepal Length(cm)')
ax.set_ylabel('Frequency')
ax.set_xlim(0,9)
ax.set_ylim(0,40)
for prop in [('major',6),('minor',3)]:
    ax.tick_params(which=prop[0],
                   axis='both',
                   direction='in',
                   length=prop[1],
                   top=True,
                   right=True)
XminorLocator = MultipleLocator(0.5)
YminorLocator = MultipleLocator(2.5)
ax.xaxis.set_minor_locator(XminorLocator)
ax.yaxis.set_minor_locator(YminorLocator)
```

 Confusions & Challenges You Faced

Learned that subplots() gives more control via ax

2. Understanding bins vs ticks
Thought bins might refer to the number of classes

Learned that:

bins = how many intervals for continuous numeric data

ticks = where the axis shows labeled values

3. Filtering by Class
Tried to filter using a string condition on a regular Python list

Got empty or incorrect results due to:

Not converting to NumPy arrays

Case sensitivity in class labels ('Iris-Setosa' vs 'Iris-setosa')

 What You Learned
How to use object-oriented plotting (fig, ax) in Matplotlib

How to use alpha for transparency and visual clarity

The purpose of bins in histograms

Importance of using NumPy arrays for filtering and vectorized operations

Always match exact spelling and case when filtering text data

How histograms show numeric distributions, not category counts

 Things to Be Cautious About
Check Your Data Before Plotting

Use print(len(...)) or print(...) to ensure you're not plotting empty arrays

Be Precise with Class Labels

Class name typos or case mismatches will silently return empty results

Use NumPy When Doing Logical Filtering

Python lists don't support == filtering like arrays do

Histogram Is for Numerical Data

Use bar plots if you want to compare counts of categories

Watch for Silent Failures

When nothing plots, check axis ranges, data lengths, or input types
